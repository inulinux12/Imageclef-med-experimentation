#### A quinto nivel

```python
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
anotaciones = pd.read_csv("anotacionesdiseaseQuinto.csv")
conceptos = pd.read_csv("conceptosDiseasesHijosQuinto.csv")
anotaciones
```

```python
conceptos
```

```python
conceptos = conceptos.set_index('Concept')
anotaciones['conceptId'] = anotaciones['conceptId'].map(conceptos['ConceptProcessed'])
anotaciones = anotaciones.drop_duplicates(['article_id','conceptId'])
anotaciones
```

```python
anotacionesgrouped = anotaciones.groupby("conceptId").count()
listafrecuencia = pd.Series(anotacionesgrouped.conceptId)
listafrecuencia.sort()
listafrecuencia
```

```python
anotacionesgrouped = anotaciones.groupby("article_id").count()
articulosfrecuencia = pd.Series(anotacionesgrouped.article_id)
articulosfrecuencia.sort()
articulosfrecuencia
```
