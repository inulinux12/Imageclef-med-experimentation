# Modality Classification

Para la clasifiación por modalidad se han usado las 30 y 31 categorías usadas en el ImageCLEF 2013. Esto se hace para garantizar que la idea de usar la estructura de Kryzhevsky permite establecer elementos diferenciativos en imágenes de dominio médico.


Las imágenes están distribuidas sobre 31 categorías las cuales se encuentran a continuación:

```python
import caffe
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
a=caffe.io.caffe_pb2.BlobProto()
file=open('caffe_models/imageclefmodality_mean.binaryproto','rb')
data = file.read()
a.ParseFromString(data)
means=a.data
means=np.asarray(means)
len(means)
means=means.reshape(3,256,256)
```

```python
MODEL_FILE = 'caffe_models/alexnet_deploy.prototxt'
PRETRAINED = 'caffe_models/alexnet_train_iter_4500.caffemodel'
IMAGE_FILE = 'caffe_models/DRXRimage.jpg'
IMAGE2 = 'caffe_models/GCHEimage.jpg'
IMAGE3 = 'caffe_models/GFLOimage.jpg'
```

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=means,
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

```python
net.set_phase_test()
net.set_mode_cpu()
```

```python
input_image = caffe.io.load_image(IMAGE3)
listimages = []
listimages.append(caffe.io.load_image(IMAGE3))
listimages.append(caffe.io.load_image(IMAGE_FILE))
listimages.append(caffe.io.load_image(IMAGE2))
plt.imshow(input_image)
```

```python
prediction = net.predict(listimages)  # predict takes any number of images, and formats them for the Caffe net automatically
print 'prediction shape:', prediction[0].shape
plt.plot(prediction[0])
#plt.plot(prediction[1])
#plt.plot(prediction[2])
print 'predicted class:', prediction[0].argmax()
```

```python
synset = pd.read_csv("Descargas/ImageCLEFTestSetGROUNDTRUTH/synset.txt", sep=" ")
synset = synset.sort("category")
synset
```

```python
groundtruth = pd.read_csv("Descargas/ImageCLEFTestSetGROUNDTRUTH/valc.txt", sep=" ", header=None, names=["filename", "class"])
#groundtruth = groundtruth.sort("class")
groundtruth
```

```python
realdict = groundtruth.set_index('filename')['class'].to_dict()
```

```python
predictedvalues = {}
input_images = []
files = sorted(realdict.keys())
for a in files:
    #print 'Descargas/ImageCLEFTestSetGROUNDTRUTH/'+a
    input_image = caffe.io.load_image('Descargas/ImageCLEFTestSetGROUNDTRUTH/'+a)
    prediction = net.predict([input_image])
    predictedvalues[a] = prediction[0].argmax()
```

```python
firstlevelrelation = pd.read_csv("RelacionModalidad.csv", sep=" ")
firstlevelrelation = firstlevelrelation.set_index('category')['firstlevel'].to_dict()
firstlevelrelation
```

```python
for a in predictedvalues.keys():
    predictedvalues[a] = firstlevelrelation[predictedvalues[a]]
    realdict[a] = firstlevelrelation[realdict[a]]
yreal, ypred = [], []
for a in predictedvalues.keys():
    yreal.append(realdict[a])
    ypred.append(predictedvalues[a])
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

Los siguientes son las convenciones de las 7 clases de primer nivel de modalidad:  
Compound or multipane images: 0  
3D reconstructions: 1  
Microscopy: 2  
Radiology: 3  
Printed signal, waves: 4  
Visible light photography: 5  
General biomedical illustrations: 6

```python
cm = confusion_matrix(yreal, ypred)

print(cm)
print "classification accuracy:", accuracy_score(yreal, ypred)
# Show confusion matrix in a separate window
plt.matshow(cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
from sklearn import metrics
yreal, ypred = [], []
for a in predictedvalues.keys():
    yreal.append(realdict[a])
    ypred.append(predictedvalues[a])


def plot_confusion_matrix(ypred, yreal):
    plt.imshow(metrics.confusion_matrix(yreal, ypred),
               cmap=plt.cm.binary, interpolation='nearest')
    plt.colorbar()
    plt.xlabel('true value')
    plt.ylabel('predicted value')
    
print "classification accuracy:", metrics.accuracy_score(yreal, ypred)
plot_confusion_matrix(yreal, ypred)
```
