# Modality Classification

Para la clasifiación por modalidad se han usado las 30 y 31 categorías usadas en el ImageCLEF 2013. Esto se hace para garantizar que la idea de usar la estructura de Kryzhevsky permite establecer elementos diferenciativos en imágenes de dominio médico.


Las imágenes están distribuidas sobre 31 categorías las cuales se encuentran a continuación:

```python
import caffe
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
a=caffe.io.caffe_pb2.BlobProto()
file=open('caffe_models/imageclefmodality_mean.binaryproto','rb')
data = file.read()
a.ParseFromString(data)
means=a.data
means=np.asarray(means)
len(means)
means=means.reshape(3,256,256)
```

```python
MODEL_FILE = 'caffe_models/alexnet_deploy.prototxt'
PRETRAINED = 'caffe_models/alexnet_train_iter_4500.caffemodel'
IMAGE_FILE = 'caffe_models/DRXRimage.jpg'
IMAGE2 = 'caffe_models/GCHEimage.jpg'
IMAGE3 = 'caffe_models/GFLOimage.jpg'
```

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=means,
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

```python
net.set_phase_test()
net.set_mode_cpu()
```

```python
input_image = caffe.io.load_image(IMAGE3)
listimages = []
listimages.append(caffe.io.load_image(IMAGE3))
listimages.append(caffe.io.load_image(IMAGE_FILE))
listimages.append(caffe.io.load_image(IMAGE2))
plt.imshow(input_image)
```

```python
prediction = net.predict(listimages)  # predict takes any number of images, and formats them for the Caffe net automatically
print 'prediction shape:', prediction[0].shape
plt.plot(prediction[0])
plt.plot(prediction[1])
plt.plot(prediction[2])
print 'predicted class:', prediction[0].argmax()
```

```python
from sklearn import metrics

def plot_confusion_matrix(y_pred, y):
    plt.imshow(metrics.confusion_matrix(y, y_pred),
               cmap=plt.cm.binary, interpolation='nearest')
    plt.colorbar()
    plt.xlabel('true value')
    plt.ylabel('predicted value')
    
print "classification accuracy:", metrics.accuracy_score(y, y_pred)
plot_confusion_matrix(y, y_pred)
```
