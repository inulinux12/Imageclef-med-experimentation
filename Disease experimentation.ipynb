# Disease experimentation

## Dataset

Proceso para generar el dataset.  
Para generar el dataset sobre diseases se escogieron los conceptos que se encontraban bajo la jerarquía de disease, se seleccionaron los artículos que fueron anotados por estos conceptos y se reanotaron con los 81 conceptos directamente hijos de disease en la ontología. De los 81 solo se usaron 71 y la siguiente es la frecuencia de aparición por artículo de cada uno de los 71 conceptos.  

```python
import caffe
from IPython.display import display, Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

A continuación cargamos los conceptos extraidos de disease.

```python
anotaciones = pd.read_csv("anotacionesdisease.csv")
conceptos = pd.read_csv("conceptosDiseases.csv")
anotaciones
```

```python
conceptos
```

A continuación se reemplazan los conceptos que jerarquicamente se encuentran bajo Disease y se reemplazan por los conceptos directamente hijos de Disease.

```python
conceptos = conceptos.set_index('Concept')
anotaciones['conceptId'] = anotaciones['conceptId'].map(conceptos['ConceptProcessed'])
anotaciones = anotaciones.drop_duplicates(['article_id','conceptId'])
anotaciones
```

La frecuencia de conceptos se puede ver acontinuación.

```python
anotacionesgrouped = anotaciones.groupby("conceptId").count()
listafrecuencia = pd.Series(anotacionesgrouped.conceptId)
listafrecuencia.sort()
listafrecuencia
```

Luego de haber hecho la anotación directa de estos conceptos sobre las imágenes, se repartieron las anotaciones un 80%-20% para entrenamiento y validación respectivamente. Vale recordar que los 71 conceptos se mapearon a una lista de conceptos de 0 a 70.

```python
train = pd.read_csv("train.txt", sep=" ")
val = pd.read_csv("val.txt", sep=" ")
train
```

```python
val
```

El número de imágenes únicas es:

```python
len(train.Imagen.unique())
```

A continuación podemos ver las imágenes que tienen mas de una anotación

```python
traingrouped = train.groupby("Imagen").count()
np.mean(traingrouped.Imagen.tolist())
```

La anterior era la media y la siguiente es la desviación estándar de anotaciones.

```python
np.std(traingrouped.Imagen.tolist())
```

```python
listafrecuencias = pd.Series(traingrouped.Imagen)
listafrecuencias.sort()
listafrecuencias
```
