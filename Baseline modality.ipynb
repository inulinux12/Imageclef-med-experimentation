```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


caffe_root = '../'  # this file is expected to be in {caffe_root}/examples

import caffe

# Set the right path to your model definition file, pretrained model weights,
# and the image you would like to classify.
MODEL_FILE = '../models/bvlc_reference_caffenet/deploy.prototxt'
PRETRAINED = '../models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'
IMAGE_FILE = 'images/cat.jpg'
```

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy'),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

```python
net.set_phase_test()
net.set_mode_gpu()
```

```python
input_image = caffe.io.load_image(IMAGE_FILE)
plt.imshow(input_image)
```

```python
prediction = net.predict([input_image], oversample=False)  # predict takes any number of images, and formats them for the Caffe net automatically
print 'prediction shape:', prediction[0].shape
plt.plot(prediction[0])
print 'predicted class:', prediction[0].argmax()
```

## Take all the images and classify them

```python
trainingpath = "/home/sebastian/Descargas/Modality/ImageCLEF2013TrainingSet/"
validationpath = "/home/sebastian/Descargas/Modality/ImageCLEFTestSetGROUNDTRUTH/"
```

```python
linestraining = np.array([line.split() for line in open(trainingpath + 'trainc.txt')])
linesvalidation = np.array([line.split() for line in open(validationpath + 'valc.txt')])
```

```python
linestraining
```

```python
linesvalidation[:, 1:]
```

```python
featurestraining = []
for a, b in linestraining:
    prediction = net.predict([caffe.io.load_image(trainingpath + a)], oversample=False)
    output = net.blobs['fc7']
    finaloutput = output.data[0].reshape((4096))
    featurestraining.append(finaloutput.copy())
    #print trainingpath + a
```

```python
featuresvalidation = []
for a, b in linesvalidation:
    prediction = net.predict([caffe.io.load_image(validationpath + a)], oversample=False)
    output = net.blobs['fc7']
    finaloutput = output.data[0].reshape((4096))
    featuresvalidation.append(finaloutput.copy())
    #print validationpath + a
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(featurestraining, linestraining[:, 1:])
```

```python
prediction = net.predict([caffe.io.load_image("/home/sebastian/Descargas/Modality/ImageCLEFTestSetGROUNDTRUTH/DVEN/1471-230X-2-15-3.jpg")], oversample=False)
output = net.blobs['fc7']
finaloutput = output.data[0].reshape((4096))
clf.predict(finaloutput)
```

```python
y_pred = clf.predict(featuresvalidation)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(linesvalidation[:, 1:].astype(int).ravel(), y_pred.astype(int))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(linesvalidation[:, 1:].astype(int).ravel(), y_pred.astype(int))

print(cm)


plt.matshow(cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
